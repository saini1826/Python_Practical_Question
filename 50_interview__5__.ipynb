```python
def is_subsequence(main_list, subseq):
    main = iter(main_list)
    return all(item in main for item in subseq)


if __name__ == "__main__":
    try:
        user_int = [2,3,4,5,6,7,8,9]
        user = [3,4,5,6]
        result = is_subsequence(user_int, user)
        print("this is a subsequence :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
# Swap using arithmetic operators

def swap_arithmetic(a, b):
    print(f"Before swap: a = {a}, b = {b}")
    a = a + b
    b = a - b
    a = a - b
    print(f"After swap: a = {a}, b = {b}")

# Swap using tuple unpacking (Pythonic way)

def swap_pythonic(a, b):
    print(f"Before swap: a = {a}, b = {b}")
    a, b = b, a
    print(f"After swap: a = {a}, b = {b}")


swap_arithmetic(5, 10)

swap_pythonic(15, 25)
```

```python
def count_nth_occurrence(s, ch, n):
    count = 0
    result = []

    for item in s:
        if item == ch:
            count += 1
            if count == n:
                continue
        result.append(item)

    return "".join(result)

if __name__ == "__main__":
    try:
        string = input("enter a string :")
        char = input('enter a str :')
        n = int(input('enter a num :'))
        result = count_nth_occurrence(string, char, n)
        print("occurence :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def first_non_repeated_char(s):
    char_count = {}
    for char in s:
        char_count[char] = char_count.get(char, 0) + 1

    for char in s:
        if char_count[char] == 1:
            return char  

text = "swiss"
result = first_non_repeated_char(text)
if result:
    print("First non-repeated character:", result)
else:
    print("No non-repeated character found.")
```

```python
def armstrong_number(lst):
    num_str = str(lst)
    num_len = len(num_str)
    number = sum(int(item) ** num_len for item in num_str)

    return number == lst

def armstrong_number_list(n):
    for item in range(1, n + 1):
        if armstrong_number(item):
            print(item, end=" , ")

if __name__ == "__main__":
    try:
        user_int = int(input("enter a number :"))
        result = armstrong_number_list(user_int)
        print("armstrong list :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finiesd")
```

```python
def generate_subsets(input_set):
    def backtrack(index, current):
        if index == len(input_set):
            subsets.append(current[:])
            return
        # Include the current element
        current.append(input_set[index])
        backtrack(index + 1, current)
        # Exclude the current element
        current.pop()
        backtrack(index + 1, current)

    subsets = []
    backtrack(0, [])
    return subsets

# Example usage
my_set = [1, 2, 3]
result = generate_subsets(my_set)
print("All subsets:")
for subset in result:
    print(subset)
```
