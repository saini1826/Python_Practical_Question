```python
def largest_word(sentence):
    words = sentence.split()
    largest = max(words, key=len)

    return largest

if __name__ == "__main__":
    try:
        user_input = input("enter a sentence :")
        result = largest_word(user_input)
        print("this is a largest word in a sentence :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def number_power(n):
    if n < 0:
        return 0
    else:
        return (n & (n - 1)) == 0

if __name__ == "__main__":
    try:
        user_input = int(input("enter a number :"))
        result = number_power(user_input)
        print("numbeer of power is :",result)
    except Exception as e:
        print("progrqam is finised")
```

```python
def flatten_a_list(lst):
    flatten = []
    for item in lst:
        if isinstance(item, list):
            flatten.extend(flatten_a_list(item))
        else:
            flatten.append(item)

    return flatten

if __name__ == '__main__':
    try:
        user_input = [1, [2, [3, 4], 5], [6, 7], 8]
        result = flatten_a_list(user_input)
        print('flatten list :',result)
    except Exception as e:
        print(e)
    finally:
        print("progra mis finised")
```

```python
def pair_list(arr, target):
    seen = set()
    pair = set()

    for item in arr:
        complement = target - item
        if complement in seen:
            pair.add(tuple(sorted((item, complement))))
        seen.add(item)

    return list(pair)

if __name__ == "__main__":
    try:
        user_input = [1,3,2,4,5,6,4,3,7,8,9,6,5]
        user = int(input('enter a target :'))
        result = pair_list(user_input, user)
        print("pair list is :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def find_median(numbers):
    nums = sorted(numbers)
    n = len(nums)
    
    mid = n // 2
    if n % 2 == 0:
        median = (nums[mid - 1] + nums[mid]) / 2
    else:
        median = nums[mid]
    
    return median

# Example usage
num_list = [5, 2, 9, 1, 7, 8]
print("Median is:", find_median(num_list))
```

```python
def sort_dictionary(data):
    sort = dict(sorted(data.items(), key=lambda x: x[1]))
    return sort

if __name__ == "__main__":
    try:
        user_input = {'banana': 3, 'apple': 5, 'cherry': 2, 'date': 4}
        result = sort_dictionary(user_input)
        print("sort dictionay :",result)
    except Exception as e:
        print(e)
    finally:
        print("progra mis finised")
```

```python
def are_rotations(str1, str2):
    if len(str1) != len(str2):
        return False
    return str2 in (str1 + str1)


string1 = input("Enter first string: ")
string2 = input("Enter second string: ")

if are_rotations(string1, string2):
    print("The strings are rotations of each other.")
else:
    print("The strings are NOT rotations of each other.")
```
