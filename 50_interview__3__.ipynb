```python
def intersection(list1, list2):
    common = set(list1) & set(list2)
    return list(common)

if __name__ == "__main__":
    try:
        user_1 = [2,3,4,5,6]
        user_2 = [7,8,3,4,9]
        result = intersection(user_1, user_2)
        print("intersection is :",result)
    except TypeError as te:
        print(te)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def remove_whitespace(string):
    remove = "".join(string.split())
    return remove

if __name__ == "__main__":
    try:
        user_input = input("enter a sentence :")
        result = remove_whitespace(user_input)
        print("without space string is :",result)
    except TypeError as te:
        print(te)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def recursive_sum(n):
    if not n:
        return 0
    else:
        return n[0] + recursive_sum(n[1:])

if __name__ == "__main__":
    try:
        user = [2,3,4,5,6]
        result = recursive_sum(user)
        print("sum of list :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def prime_number(lst):
    prime = []
    for item in range(2, lst+1):
        is_prime = True
        for i in range(2, int(item ** 0.5) + 1):
            if item % i == 0:
                is_prime = False
        if is_prime:
            prime.append(item)

    return prime

if __name__ == "__main__":
    try:
        user_input = int(input("enter a list range :"))
        result = prime_number(user_input)
        print("prime number list :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def marge_list(list1, list2):
    i , j = 0, 0
    merge = []
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merge.append(list1[i])
            i += 1
        else:
            merge.append(list2[j])
            j += 1

    merge.extend(list1[i:])
    merge.extend(list2[j:])

    return merge

if __name__ == "__main__":
    try:
        user1 = [1,3,5,7,9]
        user2 = [2,4,6,8,10]
        result = marge_list(user1, user2)
        print("merge list :",result)
    except Exception as e:
        print(e)
```

```python
def check_lst(lst):
    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            return False
    return True

if __name__ == "__main__":
    try:
        user1 = [1,3,5,7,9]
        result = check_lst(user1)
        print("merge list :",result)
    except Exception as e:
        print(e)
```

```python
def missing_num(lst):
    n = len(lst) + 1
    expect = (n * (min(lst) + max(lst))) // 2
    equal = sum(lst)
    missing = expect - equal

    return missing

if __name__ == "__main__":
    try:
        user1 = [1,2,3,5,6,7,8,9]
        result = missing_num(user1)
        print("merge list :",result)
    except Exception as e:
        print(e)
```

```python
def sum_square(n):
    return sum(i ** 2 for i in range(1,n+1))

if __name__ == "__main__":
    try:
        user_input = int(input("enter a num :"))
        result = sum_square(user_input)
        print("sum of square num :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def reverse_int(number):
    reverse = number[::-1]
    return reverse

if __name__ == "__main__":
    try:
        user_input = input("enter a num :")
        result = reverse_int(user_input)
        print("sum of square num :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def reverse_inti(lst):
    reverse = ""
    for item in lst:
        reverse = item + reverse

    return reverse

if __name__ == "__main__":
    try:
        user_input = input("enter a num :")
        result = reverse_inti(user_input)
        print("sum of square num :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```
