```python
def check_leap_year(year):
    if (year % 400 == 0 and year % 100 != 0 or year % 4 == 0):
        return True
    else:
        return False

if __name__ == "__main__":
    try:
        user_input = int(input("enter a year :"))
        result = check_leap_year(user_input)
        print("this is leap year :", result)
    except TypeError as te:
        print(te)
    except Exception as e:
        print(e)
    finally:
        print('program is finised')
```

```python
def calculate_power(base, power):
    count = 1
    for item in range(power):
        count = count * base

    return count

if __name__ == "__main__":
    try:
        user_input_ = int(input("enter a number :"))
        user_input__ = int(input('enter a nmber :'))
        result = calculate_power(user_input_, user_input__)
        print("power of base is :", result)
    except TypeError as te:
        print(te)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def genrate_even_lst(n):
    even = []
    for item in range(2,n + 1):
        if item % 2 == 0:
            even.append(item)

    return even

if __name__ == "__main__":
    try:
        user_input = int(input("enter a number :"))
        result = genrate_even_lst(user_input)
        print("even number list is :",result)
    except Exception as e:
        print(e)
    finally:
        print('program is finised')
```

```python
def reverse_words(string):
    words = string.strip().split()
    reverse = " ".join(reversed(words))

    return reverse

if __name__ == "__main__":
    try:
        user_input = input("enter a string :")
        result = reverse_words(user_input)
        print("reverse word is :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def calculate_gcd(a, b):
    while b % a != 0:
        a, b = b, a%b

    return a

if __name__ == "__main__":
    try:
        user_input1 = int(input("enter a value :"))
        user_input2 = int(input("enter a number :"))
        result = calculate_gcd(user_input1, user_input2)
        print('number od gcd is :',result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def convert_decimal_to_binary(n):
   
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //=2

    return binary

if __name__ == "__main__":
    try:
        user_input = int(input("enter a number :"))
        result = convert_decimal_to_binary(user_input)
        print("convert decimal to binary is  :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def factorial(n):
    if not n:
        return True
    else:
        return n * factorial(n - 1)

if __name__ == "__main__":
    try:
        user_input = int(input("enter a number :"))
        result = factorial(user_input)
        print("factorial is :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def largest_number(lst):
    largest = lst[0]
    for item in lst:
        if item > largest:
            largest = item

    return largest

if __name__ == "__main__":
    try:
        user_input = [2,3,4,5,6,7,8,9,10]
        result = largest_number(user_input)
        print("largest number in lst is :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def occurance(string):
    count = {}
    for item in string:
        count[item] = count.get(item, 0) + 1

    return count

if __name__ == "__main__":
    try:
        user_input = input("enter a string :")
        result = occurance(user_input)
        print("occurance is :",result)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```
